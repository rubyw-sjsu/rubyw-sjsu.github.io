Select an observation id from the result table


```
observation_ids = result['observation_id']
observation = observation_ids[0]
```

Get the associated files with the selected observation


```
observation_files = esahubble.get_associated_files(observation_id=observation)
observation_files
```

The following code loop over the file names associated to the observation to download each single file


```
for file in observation_files['filename']:
    print(f'downloading file {file}')
    esahubble.download_file(file)
```

If you are using this Notebook in ESA Datalabs, you can retrieve the files from their paths without downloading them; use the following commands to list the files paths.


```
for file in observation_files['filename']:
    print(esahubble.get_datalabs_path(file))
```
